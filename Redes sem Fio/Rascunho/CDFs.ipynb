```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Carregar o arquivo CSV
file_path = 'sa-pt-Novo.CSV'  # Substitua pelo caminho correto do seu arquivo
data = pd.read_csv(file_path)

# Colunas de interesse para a CDF
columns_of_interest = ["absErrRaw", "absErrEst", "errArtigo", "errNovoRaw"]
columns_names = ["Usando FTM", "Usando Algoritimo ESP 32", "Usando Algoritimo Artigo", "Usando Algoritimo Proposto"]

# Criar o grfico de CDF
plt.figure(figsize=(7, 5))
for column in columns_of_interest:
    data_sorted = np.sort(data[column])  # Ordenar os dados
    cdf = np.arange(1, len(data_sorted) + 1) / len(data_sorted)  # Calcular a CDF
    plt.plot(data_sorted, cdf, label=columns_names[columns_of_interest.index(column)], lw=0.5, alpha=1)

# Personalizar o grfico
plt.title("CDF dos Erros")
plt.xlabel("Erro Absoluto [m]")
plt.ylabel("CDF")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Carregar o arquivo CSV
file_path = 'sa-pt-Novo.CSV'  # Substitua pelo caminho correto do seu arquivo
data = pd.read_csv(file_path)

# Colunas de interesse para a CDF
columns_of_interest = ["absErrRaw", "absErrEst", "errArtigo", "errNovoRaw"]
columns_names = ["Usando FTM", "Usando Algoritmo ESP32", "Usando Algoritmo Artigo", "Usando Algoritmo Proposto"]

# Criar o grfico de CDF
plt.figure(figsize=(7, 5))
for column in columns_of_interest:
    hist, bin_edges = np.histogram(data[column], bins=100, density=True)  # Histograma
    cdf = np.cumsum(hist)  # Soma cumulativa
    cdf = cdf / cdf[-1]  # Normalizar para [0, 1]
    
    # Suavizao: usar os limites mdios dos bins para plotagem
    bin_centers = (bin_edges[:-1] + bin_edges[1:]) / 2
    plt.plot(bin_centers, cdf, label=columns_names[columns_of_interest.index(column)], lw=1, alpha=0.8)

# Personalizar o grfico
plt.title("CDF dos Erros")
plt.xlabel("Erro Absoluto [m]")
plt.ylabel("CDF")
plt.legend()
plt.grid(True)
plt.show()
```
